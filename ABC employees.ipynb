```python
duplicate_rows = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_rows)
```

```python
df['Height'] = np.random.randint(150, 181, size=len(df))
```

```python
most_players_team = df['Team'].value_counts().idxmax()
print("\nClub with the most number of players:", most_players_team)
min_players_team = df['Team'].value_counts().idxmin()
print("\nClub with the minimum number of players:", min_players_team)
```

```python
team_distribution = df['Team'].value_counts()
total_employees = len(df)
team_percentage = (team_distribution / total_employees) * 100
print("Distribution of employees across each team:")
print(team_distribution)
print("\n Percentage split relative to the total number of employees:")
print(team_percentage)
```

```python
plt.figure(figsize=(12, 10))
sns.countplot(x='Team', data=df, palette='magma')
plt.title('Distribution of Employees Across Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(10,6))
team_percentage.plot(kind='pie', colors = sns.color_palette('magma', len(team_percentage)))
plt.title('percentage split of Employees across Teams')
plt.ylabel('')
plt.show()
```

```python
pos_dist = df['Position'].value_counts()

print("Position Distribution:", pos_dist)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Position', data=df, palette='magma')
plt.title('Distribution of Employees Based on Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

```python
Age_bins = [20, 30, 40, 50, 60, 70]
Age_label = ['20-29', '30-39', '40-49', '50-59', '60-69']
df['Age_Group'] = pd.cut(df['Age'], bins=Age_bins, labels=Age_label, right=False)
age_group_distribution = df['Age_Group'].value_counts().sort_index()
print("Age Group Distribution:")
print(age_group_distribution)
```

```python
plt.figure(figsize=(10, 6))
age_group_distribution.plot(kind='bar', color='purple')
plt.title('Age Group Distribution of Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
team_salary_expenditure = df.groupby('Team')['Salary'].sum()
position_salary_expenditure = df.groupby('Position')['Salary'].sum()
max_team_salary = team_salary_expenditure.idxmax()
max_position_salary = position_salary_expenditure.idxmax()

print("Team with Highest Salary Expenditure:", max_team_salary)
print("Position with Highest Salary Expenditure:", max_position_salary)
```

```python
plt.figure(figsize=(10, 6))
team_salary_expenditure.plot(kind='bar', color='purple')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary ($)')
plt.show()
```

```python
position_salary_expenditure.plot(kind='bar',color='purple')
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary ($)')
plt.show()
```

```python
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
